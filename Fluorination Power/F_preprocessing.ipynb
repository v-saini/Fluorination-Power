```python

# To calclulate  Correlation and remove highly  correlated columns
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
df_raw = pd.read_excel('F260_1613.xlsx')
df_raw
```

```python
df = df_raw.drop(['Smiles', 'Solvent', ' Type of Reagent', 'Name of the Reagent', 'FPD', 'Reagent ID Number'], axis=1)
```

```python
df.head()
```

```python
df.dropna(axis=1,inplace=True)
```

```python
df.head()
```

```python
# Dropping highly correlated Features
corr_features = correlation(df, 0.80)
print("No. o features to drop : ",len(set(corr_features)))

df.drop(corr_features,axis=1,inplace=True)
```

```python
df.head()
```

```python
def remove_constant_value_features(df):
    return [e for e in df.columns if df[e].nunique() == 1]
```

```python
drop_col = remove_constant_value_features(df)
#drop_col
```

```python
new_df_columns = [e for e in df.columns if e not in drop_col]
new_df = df[new_df_columns]
new_df
```

```python
new_df[['Name of the Reagent','Smiles','Type of Reagent','Reagent ID Number', 'Solvent', 'FPD']] = df_raw[['Name of the Reagent','Smiles',' Type of Reagent','Reagent ID Number', 'Solvent', 'FPD']]
```
